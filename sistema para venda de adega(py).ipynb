```python
print("sitema de adegas para vendas")
estoque = {
    "Whisky Jack Daniels": 10,
    "Whisky Ballantines": 8,
    "Gin Gordons": 12,
    "Vodka Smirnoff": 15,
    "Cerveja Heineken": 20
}

precos = {
    "Whisky Jack Daniels": 120.0,
    "Whisky Ballantines": 90.0,
    "Gin Gordons": 70.0,
    "Vodka Smirnoff": 60.0,
    "Cerveja Heineken": 8.0
}

faturamento = 0

while True:

    print("\n===== ADEGA =====")
    print("1 - Ver estoque")
    print("2 - Registrar venda")
    print("3 - Ver faturamento")
    print("4 - Sair")

    opcao = input("Escolha uma opção: ")

    if opcao == "1":

        print("\n--- ESTOQUE ---")
        for bebida in estoque:
            print(bebida, "| Quantidade:", estoque[bebida], "| Preço: R$", precos[bebida])

    elif opcao == "2":

        nome = input("Digite o nome da bebida vendida: ")

        if nome in estoque and estoque[nome] > 0:

            estoque[nome] -= 1
            faturamento += precos[nome]

            print("Venda registrada!")

        else:
            print("Bebida não encontrada ou sem estoque!")

    elif opcao == "3":

        print("Faturamento total: R$", faturamento)

    elif opcao == "4":

        print("Sistema encerrado.")
        break

    else:
        print("Opção inválida!")
```
